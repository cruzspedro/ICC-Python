```python
def mdc(x,y):
  div1 = []
  div2 = []
  for i in range(1, x+1):
    if x % i == 0:
      div1.append(i)
  div1.reverse()
  for _ in range(1, y+1):
    if y % _ == 0:
      div2.append(_)
  div2.reverse()

  for a in div1:
    for b in div2:
      if a == b:
        return a
        break


num = int(input().strip())

expoente = int(input().strip())

print(mdc(num, expoente))

#Novamente n√£o entendi como resolver o problema recursivamente.
```
