```python
dic = []
for i in range(10):
  dic.append(float(input()))

def maior(x, y):
  if x > y:
    return x
  else:
    return y

def menor(x, y):
  if x < y:
    return x
  else:
    return y

aux = dic[0]
aux1 = dic[0]
for _ in range(9):
  aux = maior(aux, dic[_+1])
  aux1 = menor(aux1, dic[_+1])
print('%.2f'% aux1,'%.2f'% aux, sep='\n')


```
