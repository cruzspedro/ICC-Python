```python
num = int(input())

lista = lista = list(map(int,input().split()))

maior = 0

for i in range(num):
  if lista[i] > maior:
    maior = lista[i]

soma = 0

for _ in range(num):
  soma += lista[_]

media = soma/num

variancia = 0

for _ in lista:
  variancia += ((_-media)**2)/num

print("Maior: %.2f"%maior)
print("Media: %.2f"%media)
print("Variancia: %.2f"%variancia)
```
