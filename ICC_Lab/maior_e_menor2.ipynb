```python
def maior(x, y):
  if x >= y:
    return x
  else:
    return y

def menor(x, y):
  if x <= y:
    return x
  else:
    return y

entrada = input().split(' ')

lista = list(map(len,entrada))

pal_maior = lista[0]

pal_menor = lista[0]

for i in range(1,len(entrada)):
  pal_maior = maior(pal_maior,lista[i])
  pal_menor = menor(pal_menor,lista[i])

print(entrada[lista.index(pal_menor)])
print(entrada[lista.index(pal_maior)])
```
