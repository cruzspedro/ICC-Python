```python
entrada = input().split(' ')
seg = float(entrada[0])
planeta = entrada[1]
round_seg = round(seg)
def transf(x, y):
  if y == 'Terra':
    return x
  if y == 'Venus':
    return x / 243
  if y == 'Mercurio':
    return x*(176/3)
  if y == 'Jupiter':
    return x/(35760/86400)

dias = round(transf(seg, planeta)//86400)
horas = round((transf(seg, planeta) - dias*86400) // 3600)
minutos = round((transf(seg, planeta) - dias*86400 - horas*3600) // 60)
segundos = round(transf(seg, planeta) - dias*86400 - horas*3600 - minutos*60)

print(f"{round_seg} segundos no planeta {planeta} equivalem a:\n{dias} dias, {horas} horas, {minutos} minutos e {segundos} segundos")
```
