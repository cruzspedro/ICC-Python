```python
class aluno():
  def __init__(self, ID, nome, curso, idade):
    self.ID = ID
    self.nome = nome
    self.curso = curso
    self.idade = idade

  def show(self):
    print(f'Nome: {self.nome}', f'Curso: {self.curso}', f'N USP: {self.ID}', f'IDADE: {self.idade}','', sep='\n')

lista = []

while True:
  a = input().strip()
  if a == '-1':
    break
  else:
    lista.append(a)

lista2 = ['' for i in range(int(len(lista)/4))]

for i in range(len(lista2)):
  lista2[i] = list((lista[(i*4)], lista[(i*4)+1],lista[(i*4)+2],lista[(i*4)+3]))

while True:
  a = input().strip()
  if a == '-1':
    break
  elif a == '1':
    a = input().strip()
    for i in range(len(lista2)):
      if lista2[i][0] == a:
        aux = aluno(lista2[i][0],lista2[i][1],lista2[i][2],lista2[i][3])
        aux.show()
  elif a == '2':
    a = input().strip()
    for i in range(len(lista2)):
      if lista2[i][2] == a:
        aux = aluno(lista2[i][0],lista2[i][1],lista2[i][2],lista2[i][3])
        aux.show()
  elif a == '3':
    for i in range(len(lista2)):
      aux = aluno(lista2[i][0],lista2[i][1],lista2[i][2],lista2[i][3])
      aux.show() 
```
