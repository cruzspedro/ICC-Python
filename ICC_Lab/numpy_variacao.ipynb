```python
import numpy as np

num = int(input().strip())

lista = list(map(float, input().strip().split(' ')))

arr = np.array(lista)

media = 0

for i in range(num):
  media += lista[i]/num

aux = 0

for i in range(num):
  aux += ((lista[i]-media)**2)/(num-1)

desvpad = aux**0.5

cv = desvpad/np.mean(arr)

print(f'{cv:.2f}')
```
