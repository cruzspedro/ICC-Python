```python
def fatorial(x): 
  fat = 1
  i = 2
  while i <= x:
    fat = fat*i
    i = i + 1
  return fat

def anagrama(x):
  tamanho = len(x)
  denominador = 1
  aux = set(())
  for i in x:
    aux.add(i)
  aux = list((aux))
  for i in range(len(aux)):
    denominador = denominador*fatorial(x.count(aux[i]))
  return fatorial(tamanho)/denominador

num_palavras = int(input())

dic = []

for i in range(num_palavras):
  dic.append(input().strip())

for i in dic:
  print(round(anagrama(i)))
```
