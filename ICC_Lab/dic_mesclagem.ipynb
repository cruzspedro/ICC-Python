```python
entrada1 = input().strip().split(',')

num1 = list(map(int, input().strip().split(',')))

entrada2 = input().strip().split(',')

num2 = list(map(int, input().strip().split(',')))

dic1 = {}

dic2 = {}

for i in range(len(entrada1)):
  dic1.update({entrada1[i]:num1[i]})

for _ in range(len(entrada2)):
  dic2.update({entrada2[_]:num2[_]})

print('Agenda 1', dic1, sep='\n')

print('Agenda 2', dic2, sep='\n')

for andre,numero in dic1.items():
  dic2.update({andre:numero})
  
print('Agenda Mesclada', dic2, sep='\n')
```
