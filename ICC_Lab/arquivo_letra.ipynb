```python
try:
  arq = open("QuePaiseEste.txt", 'r')
except:
  print('Erro ao abrir arquivo')

lista = input().strip().split(' ')

palavra1 = lista[0]

palavra2 = lista[1]

texto = arq.read()

x = texto.replace('\n', ' ')

lista = x.split(' ')

nOcorrencia1 = lista.count(palavra1)

nOcorrencia2 = lista.count(palavra2)

print(palavra1, nOcorrencia1, palavra2, nOcorrencia2)
```
