```python
num_palavras = int(input())
dicionario = ''
for a in range(num_palavras):
  dicionario += input().strip() + ' '
palavras_dic = dicionario.split(' ')
palavra_errada = input()
for palavra in palavras_dic:

    if len(palavra) == len(palavra_errada):
        num_letras_iguais = 0

        for i in range(len(palavra_errada)):
            if palavra[i] == palavra_errada[i]:
                num_letras_iguais += 1

        porcentagem_igual = num_letras_iguais/len(palavra_errada)

        if porcentagem_igual > 0.7:
            print(palavra)


  
```
