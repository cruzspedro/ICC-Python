```python
import pickle as pck

class bebe():
  def __init__(self, data_nascimento, sexo, altura, peso, nome):
    self.data_nascimento = data_nascimento
    self.sexo = sexo
    self.altura = altura
    self.peso = peso
    self.nome = nome

  def get_data_nascimento(self):
    return (self.data_nascimento)
  
  def get_sexo(self):
    return(self.sexo)
    
  def get_altura(self):
    return (float(self.altura))   

  def get_peso(self):
    return (float(self.peso))

  def get_nome(self):
    return (self.nome)

  def set_sexo(self, sexo):
     self.__sexo = sexo

  def set_altura(self, altura):
     self.__altura = altura

  def set_peso(self, peso):
     self.__peso = peso

  def set_nome(self, nome):
     self.__nome = nome

  def set_data_nascimento(self, data_nascimento):
     self.__data_nascimento = data_nascimento

try:
  txt = open("dados-bebes.txt", 'r')
except:
  print('Erro ao abrir arquivo')

try:
  bin = open("dados-bebes.bin", 'rb')
except:
  print('Erro ao abrir arquivo')

aux = []

for i in range(30):
  aux.append(pck.load(bin, encoding='bytes'))

lista3 = [[] for i in range(int(len(aux)/5))]

for i in range(len(lista3)):
  lista3[i] = f'{aux[(i*5)]} {aux[(i*5)+1]} {aux[(i*5)+2]} {aux[(i*5)+3]} {aux[(i*5)+4]}'.split(' ')

lista1 = list(txt.readlines())

lista = [x for x in lista1]

for i in range(len(lista)):
  lista[i] = lista1[i].replace('\n', '')

lista2 = [[] for x in lista]

for i in range(len(lista)):
  lista2[i] = lista[i].split('|')

for i in range(6):
  lista2.append([])

for i in range(6):
  lista2[6+i] = lista3[i]

bebes = [[] for x in lista2]

for i in range(len(lista2)):
  bebes[i] = bebe(lista2[i][0],lista2[i][1],lista2[i][2],lista2[i][3],lista2[i][4])

totalHomens = 0

for i in range(len(bebes)):
  if lista2[i][1] == 'm':
    totalHomens += 1

mediaPesoHomens = 0

for i in range(len(bebes)):
  if lista2[i][1] == 'm':
    mediaPesoHomens += float(lista2[i][3])/totalHomens

mediaAlturaHomens = 0

for i in range(len(bebes)):
  if lista2[i][1] == 'm':
    mediaAlturaHomens += float(lista2[i][2])/totalHomens

totalMulheres = 0

for i in range(len(bebes)):
  if lista2[i][1] == 'f':
    totalMulheres += 1

mediaPesoMulheres = 0

for i in range(len(bebes)):
  if lista2[i][1] == 'f':
    mediaPesoMulheres += float(lista2[i][3])/totalMulheres

mediaAlturaMulheres = 0

for i in range(len(bebes)):
  if lista2[i][1] == 'f':
    mediaAlturaMulheres += float(lista2[i][2])/totalMulheres

print("Dados dos Bebes Homens:")
print("Total:", totalHomens)
print("Media do Peso:", f'{mediaPesoHomens:.2f}')
print("Media da Altura:", f'{mediaAlturaHomens:.2f}')
print()
print("Dados dos Bebes Mulheres:")
print("Total:", totalMulheres)
print("Media do Peso:", f'{mediaPesoMulheres:.2f}')
print("Media da Altura:", f'{mediaAlturaMulheres:.2f}')
```
