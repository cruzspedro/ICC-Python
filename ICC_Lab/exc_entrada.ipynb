```python
entrada = input().strip()

flag = True

for i in range(len(entrada)):
  if entrada[i] != '0' and entrada[i] != '1' and entrada[i] != '2' and entrada[i] != '3' and entrada[i] != '4' and entrada[i] != '5' and entrada[i] != '6' and entrada[i] != '7' and entrada[i] != '8' and entrada[i] != '9':
   flag = False

if flag == True:
  print(f"O número digitado foi: {entrada}")
else:
  print(f"Entrada Inválida.")
```
