```python
class atleta():
  def __init__(self, nome):
    self.nome = nome

class corredor(atleta):
  def __init__(self, nome):
    atleta.__init__(self, nome)
    self.correr = f'Corredor {self.nome} correndo'

class nadador(atleta):
  def __init__(self, nome):
    atleta.__init__(self, nome)
    self.nadar = f'Nadador {self.nome} nadando'

class ciclista(atleta):
  def __init__(self, nome):
    atleta.__init__(self, nome)
    self.pedalar = f'Ciclista {self.nome} pedalando'

class triatleta(corredor, nadador, ciclista):
  def __init__(self, nome):
    atleta.__init__(self, nome)
    corredor.__init__(self, nome)
    nadador.__init__(self, nome)
    ciclista.__init__(self, nome)

  def __str__(self):
    print(f'{self.correr}', f'{self.nadar}', f'{self.pedalar}', sep='\n')

nome = triatleta(input().strip())

nome.__str__()
```
