```python
entrada = input().strip().split(',')

saldo = list(map(int, input().strip().split(',')))

dic = {}

for i in range(len(entrada)):
  dic.update({entrada[i]:saldo[i]})

clientes = input().strip().split(',')

for _ in clientes:
  try:
    print(dic[_])
  except:
    print('Nao Cadastrado')
```
