```python
class produto():
  def __init__(self, codigo, preco_uni, quantidade):
    self.codigo = codigo
    self.preco_uni = preco_uni
    self.quantidade = quantidade

class nota_fiscal(produto):
  def __init__(self, codigo, preco_uni, quantidade):
    produto.__init__(self, codigo, preco_uni, quantidade)
    self.valor_total = self.preco_uni*self.quantidade

  def show(self):
    print(f'#{self.codigo:.0f}:    {self.quantidade:.0f}  {self.preco_uni:.2f} {self.valor_total:.2f}')

num_produtos = int(input().strip())

lista_aux = []

for i in range(num_produtos):
  lista_aux.append(list(map(float, input().strip().split(' '))))

num_compras = int(input().strip())

lista_aux2 = []

for i in range(num_compras):
  lista_aux2.append(list(map(float, input().strip().split(' '))))

for i in range(num_produtos):
  for a in range(num_compras):
    if lista_aux[i][0] == lista_aux2[a][0]:
      if len(lista_aux[i]) == 2:
        lista_aux[i].append(lista_aux2[a][1])
      else:
        lista_aux[i][2] += lista_aux2[a][1]

for a in lista_aux:
  if len(a) < 3:
    a.clear() 

lista = list(filter(None, lista_aux))

lista.sort()

print('#COD QTD VL_UN R$')

for i in range(len(lista)):
  aux = nota_fiscal(lista[i][0],lista[i][1],lista[i][2])
  aux.show()

total = 0

for i in range(len(lista)):
  total += lista[i][2]*lista[i][1]

print(f'Total: R$ {total:.2f}')
```
