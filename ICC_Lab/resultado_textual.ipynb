```python
num = int(input())
num2 = str(num**2)
aux = ''
for _ in range(0, len(num2)):
  if num2[_] == '0':
    aux += 'Zero '
  elif num2[_] == '1':
    aux += 'Um '
  elif num2[_] == '2':
    aux += 'Dois '
  elif num2[_] == '3':
    aux += 'Tres '
  elif num2[_] == '4':
    aux += 'Quatro '
  elif num2[_] == '5':
    aux += 'Cinco '
  elif num2[_] == '6':
    aux += 'Seis '
  elif num2[_] == '7':
    aux += 'Sete '
  elif num2[_] == '8':
    aux += 'Oito '
  elif num2[_] == '9':
    aux += 'Nove '

print(aux.strip())
```
