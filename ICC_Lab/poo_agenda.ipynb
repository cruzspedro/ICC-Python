```python
class horario():
  def __init__(self, hora, minutos, segundos):
    self.hora = hora
    self.minutos = minutos
    self.segundos = segundos

class data():
  def __init__(self, dia, mes, ano):
    self.dia = dia
    self.mes = mes
    self.ano = ano

class agenda(data, horario):
  def __init__(self, dia, mes, ano, hora, minutos, segundos, atividade):
    data.__init__(self, dia, mes, ano)
    horario.__init__(self, hora, minutos, segundos)
    self.atividade = atividade
  
  def show(self):
    print(f'{self.dia}/{self.mes}/{self.ano} - {self.hora}:{self.minutos}:{self.segundos}',f'{self.atividade}',sep='\n')

num_atividades = int(input().strip())

lista = []

for i in range(num_atividades*7):
  lista.append(input().strip())

for _ in range(num_atividades):
  aux = agenda(lista[_*7],lista[(_*7)+1],lista[(_*7)+2],lista[(_*7)+3],lista[(_*7)+4],lista[(_*7)+5],lista[(_*7)+6])
  aux.show()
```
