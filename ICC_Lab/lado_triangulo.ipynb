```python
def classificartriangulo(a, b, c):
  if a == b == c:
    print('Triangulo Equilatero')
  elif a != b and a != c and b != c:
    print('Triangulo Escaleno')
  else:
    print('Triangulo Isosceles')

def ehtriangulo(a, b, c):
  if b-c < 0:
    d = -1 * (b-c)
  else:
    d = b-c
  if a-c < 0:
    e = -1 * (a-c)
  else: e = a-c
  if a-b < 0:
    f = -1 * (a-b)
  else: f = a-b
  if not ((d < a < b+c) or (e < b < a+c) or (f < c < a+b)) :
    print('Valores nao formam um triangulo')
  else:
    classificartriangulo(a, b, c)

lado1 = int(input())
lado2 = int(input())
lado3 = int(input())

ehtriangulo(lado1, lado2, lado3)
```
