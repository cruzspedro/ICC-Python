```python
num = int(input().strip())

lista = list(map(int,input().strip().split(' ')))

flag = False

soma = 0

for _ in range(len(lista)):
  for i in range(_, len(lista)):
    soma += lista[i]
    if soma == num:
      flag = True
  soma = 0

if flag == True:
  print('SIM')
else:
  print('NAO')
```
