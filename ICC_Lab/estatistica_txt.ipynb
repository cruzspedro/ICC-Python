```python
try:
  arq = open("dados.txt", 'r')
except:
  print('Erro ao abrir arquivo')

class pessoa():
  def __init__(self, nome, sexo, idade):
    self.nome = nome
    self.sexo = sexo
    self.idade = idade

lista = arq.readlines()

entrada = []

for i in range(len(lista)):
  lista[i].replace('\n', '')

for i in range(len(lista)):
  entrada.append(list(lista[i].split(' ')))

registros = len(entrada)

homens = 0

for i in range(len(entrada)):
  if entrada[i][1] == 'm':
    homens += 1
  else:
    pass

mulheres = 0

for i in range(len(entrada)):
  if entrada[i][1] == 'f':
    mulheres += 1
  else:
    pass

media_idade = 0

for i in range(len(entrada)):
  media_idade += int(entrada[i][2])/registros

media_idade_m = 0

for i in range(len(entrada)):
  if entrada[i][1] == 'm':
    media_idade_m += int(entrada[i][2])/homens

media_idade_f = 0

for i in range(len(entrada)):
  if entrada[i][1] == 'f':
    media_idade_f += int(entrada[i][2])/mulheres

print(f'{registros} {homens} {mulheres} {media_idade:.2f} {media_idade_m:.2f} {media_idade_f:.2f}')
```
