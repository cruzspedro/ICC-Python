```python
class pessoa():
  def __init__(self, ID, nome, endereco):
    self.__ID = ID
    self.__nome = nome
    self.__endereco = endereco

  def set_ID(self, ID):
    self.__ID = ID

  def get_ID(self):
    return (self.__ID)  

  def set_nome(self, nome):
    self.__nome = nome

  def get_nome(self):
    return (self.__nome) 

  def set_endereco(self, endereco):
    self.__endereco = endereco

  def get_endereco(self):
    return (self.__endereco) 
    
class funcionario(pessoa):
  def __init__(self, ID, nome, endereco, salario_base):
    pessoa.__init__(self, ID, nome, endereco)
    self.__salario_base = float(salario_base)

  def get_salario_base(self):
    return (self.__salario_base)

  def set_salario_base(self, salario_base):
    self.__salario_base = salario_base

  def get_total_recebido(self):
    return (self.__total_recebido)

class clt(funcionario):
  def __init__(self, ID, nome, endereco, salario_base, VT, VA):
    pessoa.__init__(self, ID, nome, endereco)
    funcionario.__init__(self, ID, nome, endereco, salario_base)
    self.__VA = float(VA)
    self.__VT = float(VT)

  def set(self, VT):
      self.__VT = VT

  def set(self, VA):
      self.__VA = VA

  def get_VT(self):
    return (self.__VT)

  def get_VA(self):
    return (self.__VA)

  def get_total_recebido_clt(self):
    return (self.get_salario_base() + self.__VA + self.__VT)

  def __str__(self):
    print(f'Tipo: Clt', f'Nome: {self.get_nome()}', f'Endereco: {self.get_endereco()}',f'Salario Base: {self.get_salario_base()}',f'Transporte: {self.get_VT()}',f'Alimentacao: {self.get_VA()}',f'Total Recebido: {self.get_total_recebido_clt()}', '', sep='\n')

class comissionado(funcionario):
  def __init__(self, ID, nome, endereco, salario_base, vendas, comissao):
    pessoa.__init__(self, ID, nome, endereco)
    funcionario.__init__(self, ID, nome, endereco, salario_base)
    self.__vendas = float(vendas)
    self.__comissao = float(comissao)

  def set_vendas(self, vendas):
    self.__vendas = vendas

  def get_vendas(self):
    return (self.__vendas)

  def set_comissao(self, comissao):
    self.__comissao = comissao

  def get_comissao(self):
    return (self.__comissao)

  def get_total_recebido_comissao(self):
    return (self.get_salario_base() + (self.__comissao * self.__vendas))

  def __str__(self):
    print(f'Tipo: Comissionado', f'Nome: {self.get_nome()}', f'Endereco: {self.get_endereco()}',f'Salario Base: {self.get_salario_base()}',f'Vendas: {self.get_vendas():.0f}',f'Comissao: {self.get_comissao()}',f'Total Recebido: {self.get_total_recebido_comissao()}', '', sep='\n')

num = int(input().strip())

lista_aux = []

for i in range(num*7):
  lista_aux.append(input().strip())

lista = ['' for x in range(int(len(lista_aux)/7))]

for i in range(len(lista)):
  lista[i] = list((lista_aux[(i*7)], lista_aux[(i*7)+1], lista_aux[(i*7)+2], lista_aux[(i*7)+3], lista_aux[(i*7)+4], lista_aux[(i*7)+5], lista_aux[(i*7)+6]))

lista_aux.clear()

for i in range(len(lista)):
  lista_aux.append(int(lista[i][1]))

for i in range(len(lista)):
  lista[i][1] = lista_aux[i]

def criterio(x):
  return x[1]

lista.sort(key=criterio)

aux = ['' for x in range(len(lista))]

for i in range(len(lista)):
  if lista[i][0] == '1':
    aux[i] = clt(lista[i][1],lista[i][2],lista[i][3],lista[i][4],lista[i][5],lista[i][6])
  elif lista[i][0] == '2':
    aux[i] = comissionado(lista[i][1],lista[i][2],lista[i][3],lista[i][4],lista[i][5],lista[i][6])

for i in range(len(aux)):
  aux[i].__str__()
```
