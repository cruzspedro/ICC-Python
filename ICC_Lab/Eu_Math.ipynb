```python
mês = input()
lowercase_mês = mês.lower()
if ((lowercase_mês) == "janeiro") or ((lowercase_mês) == "março") or ((lowercase_mês) ==  "marco") or ((lowercase_mês) == "maio") or ((lowercase_mês) == "julho") or ((lowercase_mês) == "agosto") or ((lowercase_mês) == "outubro") or ((lowercase_mês) == "dezembro"):
  print('31')
elif ((lowercase_mês) == 'fevereiro'):
  print('28')
elif ((lowercase_mês) == 'abril') or ((lowercase_mês) == 'junho') or ((lowercase_mês) == 'setembro') or ((lowercase_mês) == 'novembro'):
  print('30')
else:
  print('mês invalido')
```
