```python
def dicionario(x,y,z,w):
  if len(x) == len(y): 
    dic = {}
    flag = True
    for i in range(len(x)):
      dic.update({x[i]:y[i]})
    for i in z:
      for _ in range(len(x)):
        if i == x[_]:
          return f'{i}: {dic.get(i)}\nERRO: A chave já existe no dicionário'
        else:
          pass
    if flag == True:
      for i in range(len(z)):
        dic.update({z[i]:w[i]})
    return dic
  else:
      return 'ERRO: As listas não possuem o mesmo tamanho'

nomes = input().strip().split(',')

cidades = input().strip().split(',')

nome2 = input().strip().split(',')

cidade2 = input().strip().split(',')

print(dicionario(nomes, cidades,nome2,cidade2))
```
