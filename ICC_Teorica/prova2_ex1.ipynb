```python
def imc(x,y):
  return round((float(x)/(float(y)**2)),2)

entradas = int(input().strip())

lista = []

for i in range(entradas):
  aux = []
  for a in range(5):
    aux.append(input().strip())
  lista.append(aux)

while True:
  id = input().strip()
  if id == '-1':
    break
  else:
    for _ in range(entradas):
      if lista[_].count(id) == 1:
       print(f' {lista[_][-1]}',lista[_][0],f'{imc(lista[_][2],lista[_][1]):.2f}', sep='	')
      else:
        pass
```
