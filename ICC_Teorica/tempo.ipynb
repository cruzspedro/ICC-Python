```python
tempo = int(input())

Horas = tempo // 3600
Minutos = (tempo - Horas*3600) // 60
Segundos = tempo - Horas*3600 - Minutos*60

if Horas < 10:
  Horas = f'0{Horas}'
else:
  Horas = f'{Horas}'

if Minutos < 10:
  Minutos = f'0{Minutos}'
else:
  Minutos = f'{Minutos}'

if Segundos < 10:
  Segundos = f'0{Segundos}'
else:
  Segundos = f'{Segundos}'

print(f'{Horas}:{Minutos}:{Segundos}')
```
