```python
def dicionario(x,y):
  if len(x) == len(y): 
    dic = {} 
    for i in range(len(x)):
      dic.update({x[i]:y[i]})
    return dic
  else:
    return 'ERRO: As listas não possuem o mesmo tamanho'


nomes = input().strip().split(',')

cidades = input().strip().split(',')

print(dicionario(nomes, cidades))
```
